Generating random data, graphics

```python
x = np.random.uniform(size = 50)
```

```python
y = np.random.normal(size = 50)
```

```python
% matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y);
```

```python
plt.scatter(x,y, marker = "*", color = "blue");
plt.xlabel("Random Uniform");
plt.ylabel("Random Normal");
```

```python
plt.subplot(211);
plt.scatter(x,y);
plt.subplot(212);
plt.hist(y);
```
